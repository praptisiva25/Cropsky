```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


classifier = RandomForestClassifier(n_estimators=50, max_depth=6, random_state=42)

# Train the model
classifier.fit(X_train, y_train)

# Predict on test set
y_pred_original = classifier.predict(X_test)

# Combine predictions and true labels for comparison
comparison = np.concatenate((y_pred_original.reshape(len(y_pred_original), 1),
                             y_test.reshape(len(y_test), 1)), axis=1)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_original)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Display comparison (optional)
print("Predicted vs Actual:")
print(comparison)
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
import joblib  # For loading the trained model and scaler

# Assuming your model and scaler are saved using joblib
# Load the trained model
model = joblib.load('crop_pred_97.pkl')

# Load the scaler used during training
scaler = joblib.load('scaler1.pkl')

# Load the label encoder used for the dependent variable
label_encoder = joblib.load('label_encoder1.pkl')

# Function to get valid input within the specified range
def get_valid_input(prompt, min_value, max_value):
    while True:
        try:
            # Get user input
            value = float(input(prompt))

            # Check if the value is within the valid range
            if min_value <= value <= max_value:
                return value
            else:
                print(f"Error: Value must be between {min_value} and {max_value}. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Collect user input for each feature with validation
N = get_valid_input("Enter the value for Nitrogen (N) [0-140]: ", 0, 140)
P = get_valid_input("Enter the value for Phosphorus (P) [5-145]: ", 5, 145)
K = get_valid_input("Enter the value for Potassium (K) [5-205]: ", 5, 205)
temperature = get_valid_input("Enter the temperature in Celsius [8-44]: ", 8, 44)
humidity = get_valid_input("Enter the humidity percentage [14-100]: ", 14, 100)
ph = get_valid_input("Enter the soil pH value [3-11]: ", 3, 11)
rainfall = get_valid_input("Enter the rainfall in mm [20-300]: ", 20, 300)

# Combine these into a single array
user_input = np.array([[N, P, K, temperature, humidity, ph, rainfall]])

# Scale the input features using the same scaler
user_input_scaled = scaler.transform(user_input)

# Predict using the trained model
encoded_prediction = model.predict(user_input_scaled)

# Decode the prediction to get the original label
prediction_label = label_encoder.inverse_transform(encoded_prediction.astype(int))

# Output the prediction
print(f"The recommended crop for the given conditions is: {prediction_label[0]}")
```
